分析の流れ(因果推論)  
1. ビジュアルで、全体の変数の特徴を掴む
2. 全体の変数を把握して、変数をカテゴリ変数と連続変数に分ける。さらに、カテゴリ変数は名義尺度と順序尺度に分ける。（リストあるいは辞書に格納するのを忘れずに!）
3. 目的変数と相関が高い説明変数について調べる
4. 欠損値・異常値の処理
5. 新しい説明変数(特徴量)の生成
6. 回帰問題の場合は、歪んだデータを正規分布に従うデータに変換（対数変換など）する。*必ずしも行う必要はない。
7. 適切なアルゴリズムの選択

回帰（Regression）分析の場合、ターゲットの値がどちらかに偏らず正規分布に従っていることが重要  
やはり正規分布になっていない。  
目的変数の残差の分布は正規分布になっていないとモデルの予測精度に影響が出てしまうので、これはあとで対数変換する必要があるかもしれない。  
回帰分析の前提は4つある。
1. 独立性（データそれぞれが独立）
2. 等分散性（説明変数にかかわらず分散が一定）
3. 正規性（誤差自体が正規分布している）
4. 線形性（説明変数と目的変数の関係は直線で近似できる）

１．独立性、２．等分散性、３．正規性の3つは、残差が満たす必要がある。  
残差とは、目的変数の実際の値と、回帰式で計算された予測値の差を言う。この残差を、誤差の代わりに使っている。誤差は、目的変数の実際の値と、真の回帰式で計算される目的変数の真の値の差。真の値は推測しかできないため、真の誤差は計算できない。それゆえ、残差を使うわけである。  
残差プロットに特異なパターンがなければ、１．独立性は満たされる。  
残差が均一ならば、２．等分散性の仮定は満たされる。

```python
#歪度と尖度を計算
print("歪度: %f" % train['SalePrice'].skew())
print("尖度: %f" % train['SalePrice'].kurt())
```

歪度が正の値なので、右裾が長い（≒左に偏っている）分布であること、さらに尖度が正の値のため正規分布よりもだいぶ尖った（平均付近にデータが集中している）分布であることがわかる。

```python
#分析方針を立てるために、何の変数が「SalePrice（住宅価格）」と強い相関があるが一気に確認
corrmat = train.corr()
# ヒートマップに表示させるカラムの数
k = 11

# SalesPriceとの相関が大きい上位10個のカラム名を取得
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index

# SalesPriceとの相関が大きい上位10個のカラムを対象に相関を算出
cm = np.corrcoef(train[cols].values.T)

# ヒートマップのフォントサイズを指定
fig, ax = plt.subplots(figsize=(12, 10))
sns.set(font_scale=1.2)

# 算出した相関データをヒートマップで表示
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt=".2f", annot_kws={"size": 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
fig.savefig("figure4.png")
```

この結果から以下の点が読み取れる

SalePrice（住宅価格）と強い相関があるのは以下
1. OverallQual：全体的な素材と仕上げの品質(1~10)
2. GrLivArea：リビングの広さ(平方フィート)
3. GarageCars：車の容量に対する車庫のサイズ(0~4)
4. GarageArea：車庫サイズ(平方フィート)
5. TotalBsmtSF：地下の広さ(総平方フィート)
6. 1stFlrSF：1階の広さ(平方フィート)
7. FullBath：グレード以上のフルバスルーム(0~3)
8. TotRmsAbvGrd:地上階の部屋数(浴室除く)
9. YearBuilt：建設年
10. YearRemodAdd:改造年(改造していない場合、YearBuiltと同じ)

強い相関がある（多重共線性の点から一つに絞る必要がある）
→　多重共線性とは、説明変数間の相関が高いために、有意水準(t値)が期待より非常に低くなり、有意な推定結果が得られないような結果が生じる現象  
    (本来、説明変数と目的変数に統計的に関係があるのに、関係なしとして出力されてしまう可能性がある)

・GarageCars：車の容量に対する車庫のサイズ(0~4)  
・GarageArea：車庫サイズ(平方フィート)  


・TotalBsmtSF：地下の広さ(総平方フィート)  
・1stFlrSF：1階の広さ(平方フィート)  


・GrLivArea：リビングの広さ(平方フィート)  
・TotRmsAbvGrd:地上階の部屋数(浴室除く)  


「GarageCars」「GarageArea」などは項目の意味からも相関の強さは明らかで、これらは多重共線性を引き起こす。  
※ 最も一般的な解消法は、「相関関係が高いと考えられる説明変数のいずれかを外すこと」
今回は、回帰分析の説明変数から「GarageCars」、「TotRmsAbvGrd」を外す
このような状況を検出するのにヒートマップは便利。  

住宅の広さを表す変数はそれぞれ階ごとに分かれているため、合計して変数を新規に作成する  
「1stFlrSF：1階の広さ(平方フィート)」 + 「2ndFlrSF：2階の広さ(平方フィート)」 + 「TotalBsmtSF：地下の広さ(総平方フィート)」

```python
train["TotalSF"] = train["1stFlrSF"] + train["2ndFlrSF"] + train["TotalBsmtSF"] 
```

```python
#住宅の広さと住宅価格の関係の散布図、回帰分析
fig = plt.figure(figsize=(9,6))
sns.regplot(x=train["TotalSF"], y=train["SalePrice"])
plt.xlabel("TotalSF(住宅の広さ合計)"")
plt.ylabel("SalePrice(住宅価格)")
plt.show()
```

外れ値の値あり、「TotalSF」が6500以上のデータ除去する

```python
#外れ値を除外する
train = train.drop(train[(train['TotalSF']>6500)].index)
```

```python
#再度、散布図
fig = plt.figure(figsize=(9,6))
sns.regplot(x=train["TotalSF"], y=train["SalePrice"])
plt.xlabel("TotalSF(住宅の広さ合計)")
plt.ylabel("SalePrice(住宅価格)")
plt.show()
```

```python
#単回帰分析
#statsmodelsでは、定数項の追加は自動で行われないため、「sm.add_constant([説明変数])」を使って明示的に追加する必要あり
#sm.OLS(endog=[目的変数], exog=[説明変数])
explanatory_variable = sm.add_constant(train["TotalSF"])
model = sm.OLS(
    endog=train["SalePrice"],
    exog=explanatory_variable
)
result = model.fit()
result.summary()
```

推定結果  
Y = -3274 + 83.4436X  
説明変数の係数は、Xが1単位変化した時のYの変化量を示す  
つまり、TotalSF(住宅の広さ合計)が1単位増えると、SalePrice(住宅価格)が83.4436上がることが予測できる  
「std err」、「t」、「P>|t|」はそれぞれ、標準誤差、t値、p値と呼ばれる統計量を表す。  
標準誤差は、対応する回帰係数の標準偏差、つまり推定精度を表す。  
t値、p値は有意水準をはかるための指標。  
t値は、係数に対する相対的な誤差を示す。t値 = 説明変数の係数 / 標準誤差
p値は、回帰係数がゼロ(目的変数に与える効果なし)と仮定した時にこのデータが得られる確率。  
一般的に、0.05(5%)を基準として、その値を下回っていれば、統計的に有意に効果があると言える。  
例えば、p値=0.05の場合、回帰係数がゼロである確率が5% → 95%の確率でゼロではない →　該当、説明変数は目的変数に係数分の影響を与える

```python
#推定結果をもとに、住宅価格の予測
#定数項=1、TotalSF(住宅の広さ合計)=3000の場合のSalePrice(住宅価格)の値
result.predict([1, 3000])
```

```python
#重回帰分析
x = train[['OverallQual', 'GarageArea', 'FullBath', 'TotalSF']]
y = train[["SalePrice"]]
explanatory_variable = sm.add_constant(x)
model = sm.OLS(y, explanatory_variable)
result = model.fit()
result.summary()
```

```python
#単回帰分析
#FullBath(バスルームのグレード(0~3))
#単回帰分析では、有意に正の効果あり
#重回帰分析では、有意性なくなる
explanatory_variable = sm.add_constant(train["FullBath"])
model = sm.OLS(
    endog=train["SalePrice"],
    exog=explanatory_variable
)
result = model.fit()
result.summary()
```

```python
#重回帰分析
#FullBathまだ有意
x = train[['OverallQual', 'FullBath']]
y = train[["SalePrice"]]
explanatory_variable = sm.add_constant(x)
model = sm.OLS(y, explanatory_variable)
result = model.fit()
result.summary()
```
